```python
import pandas as pd
```

```python
df = pd.read_csv('EU-Popular-Cities-Prices - PreProcesed.csv', sep=',')
df
```

```python
# Filter price items to include in basic costs

filter = df.item_id.isin([1,4,40,44,8,9,10,11,12,13,19,110,111,112,113,115,116,118,119,121,120,20,30,33,27,105])
df_filtered = df[filter]
```

```python
# Dictionaries for conversion and bining purposes

toMonthlyCostDict = {1:4, 4:4, 40:1, 44:1, 8:4, 9:4, 10:4, 11:4, 12:4, 13:4, 19:4, 110:4, 111:4, 112:4, 113:4, 115:4, 116:4, 118:4, 
                      119:4, 121:4, 20:1, 30:1, 33:1, 27:1, 105:1}
itemType = {1:'HOBBIES', 4:'HOBBIES', 40:'HOBBIES', 44:'HOBBIES', 8:'MARKET', 9:'MARKET', 10:'MARKET', 11:'MARKET', 12:'MARKET',
                13:'MARKET', 19:'MARKET', 110:'MARKET', 111:'MARKET', 112:'MARKET', 113:'MARKET', 115:'MARKET', 116:'MARKET', 
                118:'MARKET', 119:'MARKET', 121:'MARKET', 20:'TRANSPORT', 30:'UTILITIES', 33:'UTILITIES', 27:'RENT', 105:'SALARY'}
```

```python
# Appending new columns to dataframe
a = []
b = []
for itemId in df_filtered.item_id:
    a.append(toMonthlyCostDict[itemId])
    b.append(itemType[itemId])
df_filtered['toMonthlyCost'] = a
df_filtered['item_type'] = b
```

```python
df_filtered.head()
```

```python
# Calculating monthly prices for each iterm

df_filtered['avg_monthly_cost'] = df_filtered.average_price * df_filtered.toEur * df_filtered.toMonthlyCost 
data = df_filtered[['city_id', 'city_name', 'item_id', 'item_name', 'item_type', 'avg_monthly_cost']]
data.head()
```

```python
# Aggregating items by category 

data = data.groupby(['city_name', 'item_type']).sum()
data[:20]
```

```python
# Aggregation Output -> using rest_index() to undo groupBy binding on (city_name, item_type)
data = data.reset_index()[['city_name','item_type','avg_monthly_cost']]
data
```

```python
# Total cost of living calculation
cashflow_ne_salary = data['item_type'] != 'SALARY'
total_cost_df = data[cashflow_ne_salary].groupby(['city_name']).sum() # SUMS ALL AVG_MONTHLY_PRICES FOR EACH NON-SALARY-ITEM
total_cost_df = total_cost_df.reset_index()
total_cost_df = total_cost_df.rename(columns={'avg_monthly_cost':'avg_total_cost_of_living'})
total_cost_df
```

```python
#Pivot dataframe
data_pivoted = data.pivot(index = 'city_name', columns = 'item_type', values = 'avg_monthly_cost')
data_pivoted
```

```python
data = pd.DataFrame(data_pivoted.to_records())
data
```

```python
data[['CITY','COUNTRY']] = data.city_name.str.split(',', expand=True)
data
```

```python
#Rearang columns
data = data[['city_name','CITY','COUNTRY','HOBBIES','MARKET','RENT','TRANSPORT','UTILITIES','SALARY']]
```

```python
data['TOTAL EXPENSES'] = data.HOBBIES + data.MARKET + data.RENT + data.TRANSPORT + data.UTILITIES
data['TOTAL SAVINGS'] = data.SALARY - data['TOTAL EXPENSES']
```

```python
data['COUNTRY'] = data['COUNTRY'].str.lstrip() #Remove blanks from split
data = data.rename(columns={'city_name':'COMPLETE CITY NAME'})
data
```

```python
data.to_csv('EU-Popular-Cities-Summarized.csv', sep=',', index=False)
```
